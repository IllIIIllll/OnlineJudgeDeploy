```python
n = list(map(int, input().split()))

sum([i + 1 for i in range(max(n))]) - sum(n)
```

```python
import random

# Create test case input
num_list = [i+1 for i in range(1, 10000)]
tc_in = [[i+1 for i in range(random.choice(num_list))] for _ in range(100)]
for i in tc_in:
    random.shuffle(i)
    i.pop(random.choice(i)-1)

# Write test case
for i in range(100):
    file_in = open(f"D:/JiSung/algorithm/6. 1에서 n 사이의 수 중 없는 수 찾기/datasets/{i+1}.in", 'w')
    file_out = open(f"D:/JiSung/algorithm/6. 1에서 n 사이의 수 중 없는 수 찾기/datasets/{i+1}.out", 'w')
    
    file_in.write(' '.join(map(str, tc_in[i])))
    file_out.write(f'{sum([i + 1 for i in range(max(tc_in[i]))]) - sum(tc_in[i])}')
    
    file_in.close()
    file_out.close()
```

```python
tc_in = [max(i) for i in tc_in]
tc_in_ = set(tc_in)
tc_in == tc_in_
```
